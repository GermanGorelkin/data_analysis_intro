# Сортировка массивов

```python
import numpy as np
```

```python
x = np.array([2, 1, 4, 3, 5])
np.sort(x)
```

```python
x.sort()
print(x)
```

```python
# индексы отсортированных элементов
x = np.array([2, 1, 4, 3, 5])
i = np.argsort(x)
print(i)
```

```python
x[i]
```

### Сортировка по строкам и столбцам

```python
rand = np.random.RandomState(42)
X = rand.randint(0, 10, (4, 6))
print(X)
```

```python
np.sort(X, axis=0)
```

```python
np.sort(X, axis=1)
```

### Секционирование

```python
x = np.array([7, 2, 3, 1, 6, 5, 4])
np.partition(x, 3)
```

```python
np.partition(X, 2, axis=1)
```
