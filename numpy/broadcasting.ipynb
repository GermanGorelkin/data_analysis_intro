### Пример 1
Сложение одномерного массива с двумерным

```python
M = np.ones((2, 3))
a = np.arange(3)
print(M)
print(b)
```

```python
print(M.shape)
print(b.shape)
```

Правило 1  
По сколько размерность массива a меньше, дополняем его измерением слева  
M.shape -> (2, 3)  
a.shape -> (1, 3)

Правило 2  
Первое измерение массивов различаются, по этому растягиваем его до совподения  
M.shape -> (2, 3)  
a.shape -> (2, 3)

```python
M + a
```

### Пример 2
Траслирование двух масиивов

```python
a = np.arange(3).reshape((3, 1))
b = np.arange(3)
print(a)
print(b)
```

```python
print(a.shape)
print(b.shape)
```

Привило 1  
Дополняем b еденицами  
a.shape -> (3, 1)  
b.shape -> (1, 3)

Правило 2  
Добавляем еденицы до совподения  
a.shape -> (3, 3)  
b.shape -> (3, 3)

```python
a + b
```

## Центрирование массива

```python
X = np.random.random((10, 3))
X
```

```python
# среднее
Xmean = X.mean(0)
Xmean
```

```python
# центрирование 
X_centered = X - Xmean
X_centered
```

```python
X_centered.mean(0)
```

## График двумерной функции

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)[:, np.newaxis]

z = np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.imshow(z, origin='lower', extent=[0, 5, 0, 5],
           cmap='viridis')
plt.colorbar();
```
