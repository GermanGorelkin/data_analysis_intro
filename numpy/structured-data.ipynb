# Структурированные данные

```python
import numpy as np
```

```python
data = np.zeros(4, dtype={'names':('name', 'age', 'weight'),
                          'formats':('U10', 'i4', 'f8')})
print(data.dtype)
```

```python
name = ['Alice', 'Bob', 'Cathy', 'Doug']
age = [25, 45, 37, 19]
weight = [55.0, 85.5, 68.0, 61.5]
```

```python
data['name'] = name
data['age'] = age
data['weight'] = weight
print(data)
```

```python
data['name']
```

```python
data[0]
```

```python
data[data['age'] < 30]['name']
```

## Создание струк структурированных массивов

```python
np.dtype({'names':('name', 'age', 'weight'),
          'formats':('U10', 'i4', 'f8')})
```

```python
np.dtype({'names':('name', 'age', 'weight'),
          'formats':((np.str_, 10), int, np.float32)})
```

```python
np.dtype([('name', 'S10'), ('age', 'i4'), ('weight', 'f8')])
```

```python
np.dtype('S10,i4,f8')
```
